Q1

```python
class Solution:
    def runningSum(self, nums: List[int]) -> List[int]:
        result=[]
        result.append(nums[0])
        for i in range(1,len(nums)):
            result[i]=result[i-1]+nums[i]
        return result
```

Q2

```python
class Solution:
    def shuffle(self, nums: List[int], n: int) -> List[int]:
        r1=[]
        r2=[]
        result=[]
        for i in range(0,n):
            r1.append(nums[i])
        for i in range(n,2*n):
            r2.append(nums[i])
        for i in range(n):
           result.append(r1[i])
           result.append(r2[i])
           
        return result
        
```

Q3

```python
class Solution:
    def kidsWithCandies(self, candies: List[int], extraCandies: int) -> List[bool]:
        n=len(candies)
        list=[]
        m=0
        for i in range(0,n):
            if(candies[i]>m):
                m=candies[i]
        for i in range(0,n):
            if(candies[i]+extraCandies>=m):
                list.append(True)
            else:
                list.append(False)
        return list
```
