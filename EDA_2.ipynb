Q1

```python
class Solution:
    def defangIPaddr(self, address: str) -> str:
        result=""
        for i in address:
            if i==".":
                result=result+"[.]"
            else:
                result=result+i
        return result
```

Q2

```python
class Solution:
    def findNumbers(self, nums: List[int]) -> int:
        ogcount=0
        # incount=0
        for i in nums:
            incount=0
            for j in str(i):
                incount+=1
            if incount%2==0:
                ogcount+=1
        return ogcount

```

class Solution:
    def numIdenticalPairs(self, nums: List[int]) -> int:
        count=0
        for i in range (0,len(nums)):
            for j in range(i+1,len(nums)):
                if nums[i]==nums[j]:
                    count+=1
        return count


        

Q4


```python
class Solution:
    def smallerNumbersThanCurrent(self, nums: List[int]) -> List[int]:
        result=[]
        for i in range(0,len(nums)):
            count=0
            for j in range(0,len(nums)):
                if i!=j:
                    if nums[j]<nums[i]:
                        count+=1
            result.append(count)
        return result
```

Q5

```python
class Solution:
    def subtractProductAndSum(self, n: int) -> int:
        pr=1
        s=0
        for i in str(n):
            pr=pr*int(i)
            s=s+int(i)
        return pr-s
```

Q6

```python
class Solution:
    def xorOperation(self, n: int, start: int) -> int:
        nums=[]
        for i in range(0,n):
            nums.append(start+2*i)
        result=0
        for i in range(0,n):
            result=result^nums[i]
        return result
        
        
```
